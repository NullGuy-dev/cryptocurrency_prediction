```python
import numpy as np
import pandas as pd
from itertools import cycle
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv("/content/drive/MyDrive/coin_Ethereum.csv")
data.head()
```

```python
pred_period = 5
data["Predicted"] = data[["Close"]].shift(-pred_period)
data.head()
```

```python
scaler = MinMaxScaler()
colm = ["High", "Low", "Open", "Close", "Volume", "Marketcap", "Predicted"]
scaler.fit(data[colm])
data[colm] = scaler.transform(data[colm])
data.head()
```

```python
X = np.array(data[["High", "Low", "Open", "Close", "Volume", "Marketcap"]])
X = X[:-pred_period]
```

```python
y = data["Predicted"].values
y = y[:-pred_period]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
rf_reg = RandomForestRegressor(max_depth=7)
rf_reg.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
pred = rf_reg.predict(X_test)
r2 = r2_score(pred, y_test)
mse = mean_squared_error(pred, y_test)
print(f"R2-score: {r2:.3f}")
print(f"MSE: {mse}")
```
